```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/usr/u_food_marketing.csv')
df

pd.set_option('display.max_rows', 3000)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df[df.duplicated()].count()
df.drop_duplicates(keep=False, inplace=True)
pd.set_option('display.max_columns', 50)
df
df.head()
```

```python
df['marital_Divorced'] = df['marital_Divorced'].replace({1:1, 0:0})
df['marital_Married'] = df['marital_Married'].replace({1:2, 0:0})
df['marital_Single'] = df['marital_Single'].replace({1:3, 0:0})
df['marital_Together'] = df['marital_Together'].replace({1:4, 0:0})
df['marital_Widow'] = df['marital_Widow'].replace({1:5, 0:0})

df['marital_stat'] = df['marital_Divorced'] + df['marital_Married'] + df['marital_Single'] + df['marital_Together'] + df['marital_Widow']
df['marital_status'] = df['marital_stat'].replace({1:'Divorced', 2:'Married', 3:'Single', 4:'Together', 5:'Widow'})
df[df['marital_status'] != 0].head()

df.drop(['marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together', 'marital_Widow', 'marital_stat'], axis=1, inplace=True)
df.head()
```

```python
df['Children'] = df['Kidhome'] + df['Teenhome']
df.head()
df.drop(['Kidhome','Teenhome'], axis=1, inplace=True)
df.head()
```

```python
df['Total_campaigns'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] + df['Response']
df.drop(['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response'], axis=1, inplace=True)
df.head()
```

```python
df['education_2n Cycle'] = df['education_2n Cycle'].replace({1:1, 0:0})
df['education_Basic'] = df['education_Basic'].replace({1:2, 0:0})
df['education_Graduation'] = df['education_Graduation'].replace({1:3, 0:0})
df['education_Master'] = df['education_Master'].replace({1:4, 0:0})
df['education_PhD'] = df['education_PhD'].replace({1:5, 0:0})
df['education_status'] = df['education_2n Cycle'] + df['education_Basic'] + df['education_Graduation'] + df['education_Master'] + df['education_PhD']
df['education_status'] = df['education_status'].replace({1:'2n Cycle', 2:'Basic', 3:'Graduation', 4:'Master', 5:'PhD'})
df.drop(['education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD'], axis=1, inplace=True)
df.head()
```

```python
df.info()
df['Total_campaigns'].sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.select_dtypes(include=['object']).columns:
  df[col] = le.fit_transform(df[col])
```

```python
correlation = df.corr(method='pearson')['Total_campaigns'].sort_values(ascending=False)
correlation

#using an heatmap to visualize the correlation of data with total campaigns

sns.heatmap(df.corr(method='pearson'), annot=False, cmap='coolwarm')
plt.show()
```

```python
all_corr = df.corr(method='pearson')
all_corr = all_corr[(all_corr >= 0.3) & (all_corr < 1)]

sns.heatmap(all_corr)
```

```python
age_groups = [(23,30), (31,40), (41,50), (51,60), (61,70), (71,85)]

#Defining a function to assign every person to the right age group
def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return("Unknown")

df['Age_Group'] = df['Age'].apply(assign_age_group)
```

```python
df[['Age','Age_Group']].head()
```

```python
age_order = ['23 - 30', '31 - 40', '41 - 50', '51 - 60', '61 - 70', '71 - 85']
sns.pointplot(data = df, x = 'Age_Group', y = 'Total_campaigns', order = age_order)
```

```python
count = df['Age_Group'].value_counts()
percentage = count / df.shape[0] * 100
percentage
```

```python
percentage_df = percentage.reset_index()
percentage_df.columns = ['Age_Group', 'Percentage']
percentage_df = percentage_df.sort_values('Age_Group')
percentage_df
```

```python
sns.barplot(data = percentage_df, x = 'Age_Group', y = 'Percentage')
plt.title('Percentage of Total Campaigns by Age Groups')
plt.show()
```
